```python
# Load diabetes dataset from sklearn
from sklearn.datasets import load_diabetes
import pandas as pd

# Alternative - use seaborn built-in dataset
import seaborn as sns
df = sns.load_dataset('healthexp')
print("Dataset loaded!")
print("Shape:", df.shape)
print(df.head())
```

```python
# Explore the dataset
print("Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nCountries in dataset:")
print(df['Country'].unique())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Chart 1 - Healthcare Spending by Country
spending_by_country = df.groupby('Country')['Spending_USD'].mean().reset_index()
spending_by_country = spending_by_country.sort_values('Spending_USD', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(spending_by_country['Country'],
        spending_by_country['Spending_USD'],
        color='steelblue')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Healthcare Spending (USD)', fontsize=12)
plt.title('Average Healthcare Spending by Country\nPopulation Health Analysis', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('spending_by_country.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart 1 saved!")
```

```python
# Chart 2 - Life Expectancy by Country
life_exp_country = df.groupby('Country')['Life_Expectancy'].mean().reset_index()
life_exp_country = life_exp_country.sort_values('Life_Expectancy', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(life_exp_country['Country'],
        life_exp_country['Life_Expectancy'],
        color='mediumseagreen')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Life Expectancy (Years)', fontsize=12)
plt.title('Average Life Expectancy by Country\nPopulation Health Analysis', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('life_expectancy_by_country.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart 2 saved!")
```

```python
plt.figure(figsize=(10, 6))
countries = df['Country'].unique()
colors = ['steelblue', 'coral', 'mediumseagreen', 'purple', 'orange', 'red']
for i, country in enumerate(countries):
    country_data = df[df['Country'] == country]
    plt.scatter(country_data['Spending_USD'], country_data['Life_Expectancy'], label=country, color=colors[i], s=50)
plt.xlabel('Healthcare Spending (USD)', fontsize=12)
plt.ylabel('Life Expectancy (Years)', fontsize=12)
plt.title('Healthcare Spending vs Life Expectancy', fontsize=14)
plt.legend()
plt.tight_layout()
plt.savefig('spending_vs_life_expectancy.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart 3 saved!")
```

```python
plt.figure(figsize=(12, 6))
for i, country in enumerate(countries):
    country_data = df[df['Country'] == country]
    plt.plot(country_data['Year'],
             country_data['Spending_USD'],
             label=country,
             color=colors[i],
             linewidth=2)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Healthcare Spending (USD)', fontsize=12)
plt.title('Healthcare Spending Trends Over Time\nPopulation Health Analysis', fontsize=14)
plt.legend()
plt.tight_layout()
plt.savefig('spending_trend_over_time.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart 4 saved!")
```
