```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load using seaborn built-in healthcare dataset
# For Medicare analysis we'll use CMS data structure
df = sns.load_dataset('healthexp')

# Create SQLite database
conn = sqlite3.connect('healthcare.db')
df.to_sql('healthcare', conn, if_exists='replace', index=False)

print("Database created!")
print("Shape:", df.shape)
print(df.head())
```

```python
# SQL Query 1 - Top 5 Countries by Average Healthcare Spending
query1 = """
SELECT Country,
       ROUND(AVG(Spending_USD), 2) as Avg_Spending,
       ROUND(AVG(Life_Expectancy), 2) as Avg_Life_Expectancy
FROM healthcare
GROUP BY Country
ORDER BY Avg_Spending DESC
"""
result1 = pd.read_sql_query(query1, conn)
print("Top Countries by Healthcare Spending:")
print(result1)
```

```python
# SQL Query 2 - Spending Growth by Decade
query2 = """
SELECT Country,
       CASE
           WHEN Year BETWEEN 1970 AND 1979 THEN '1970s'
           WHEN Year BETWEEN 1980 AND 1989 THEN '1980s'
           WHEN Year BETWEEN 1990 AND 1999 THEN '1990s'
           WHEN Year BETWEEN 2000 AND 2009 THEN '2000s'
           WHEN Year BETWEEN 2010 AND 2019 THEN '2010s'
           ELSE '2020s'
       END as Decade,
       ROUND(AVG(Spending_USD), 2) as Avg_Spending
FROM healthcare
GROUP BY Country, Decade
ORDER BY Country, Decade
"""
result2 = pd.read_sql_query(query2, conn)
print("Healthcare Spending by Decade:")
print(result2)
```

```python
# SQL Query 3 - Most Efficient Countries
# (High Life Expectancy relative to Spending)
query3 = """
SELECT Country,
       ROUND(AVG(Spending_USD), 2) as Avg_Spending,
       ROUND(AVG(Life_Expectancy), 2) as Avg_Life_Expectancy,
       ROUND(AVG(Life_Expectancy) / AVG(Spending_USD) * 1000, 4) as Efficiency_Score
FROM healthcare
GROUP BY Country
ORDER BY Efficiency_Score DESC
"""
result3 = pd.read_sql_query(query3, conn)
print("Healthcare Efficiency by Country:")
print("(Life Expectancy gained per $1000 spent)")
print(result3)
```

```python
# SQL Query 4 - Recent vs Historical Spending Comparison
query4 = """
SELECT Country,
       ROUND(AVG(CASE WHEN Year <= 1990 THEN Spending_USD END), 2) as Historical_Avg,
       ROUND(AVG(CASE WHEN Year > 1990 THEN Spending_USD END), 2) as Recent_Avg,
       ROUND(AVG(CASE WHEN Year > 1990 THEN Spending_USD END) -
             AVG(CASE WHEN Year <= 1990 THEN Spending_USD END), 2) as Spending_Increase
FROM healthcare
GROUP BY Country
ORDER BY Spending_Increase DESC
"""
result4 = pd.read_sql_query(query4, conn)
print("Historical vs Recent Spending:")
print(result4)
```
